```python
with open("../23.02_AP1/2_Input/thucydides.txt", 'r', encoding="utf-8") as fp:
    file_string = fp.read()

list_words = file_string.split()
# print(len(list_words), list_words[:20])
```

```python
list_cleanwords = []
for word in list_words:
    word_strip = word.strip("""-?,().!":;'""")
    if word_strip.isalpha():
        list_cleanwords.append(word_strip.lower())

# print(len(list_cleanwords))
```

```python
word_count = {}
for word in list_cleanwords:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# print(len(word_count))
```

```python
max_used = 0
max_used_word = ""
for word in word_count:
    #print(word, word_count[word])
    if word_count[word] > max_used:
        max_used = word_count[word]
        max_used_word = word

print(max_used, max_used_word)
```

```python
count_word = []
for n in range(max_used+1):
    count_word.append([])

for word, count in word_count.items():
    count_word[int(count)].append(str(word))

count = 0
m = max_used
while m > 0:
    for x in count_word[m]:
        print(x, "=", m)
        count = count + 1
    m = m - 1
    if count >= 10:
        break
```

```python
word_count_list = []
for n in range(max_used+1):
    word_count_list.append([])

for word, count in word_count.items():
    word_count_list[int(count)].append(str(word))
    
print(word_count_list)
```

```python
len_ = 0
len_str = []

# 검산용
# for x in word_count:
#     if len(x) == 15:
#         len_str.append(x)

for x in word_count:
    if len_ < len(x):
        len_str.clear()
        len_str.append(x)
        len_ = len(x)
    elif len_ == len(x):
        len_str.append(x)

print(len_, len_str)
```

```python
sum_ = 0
strs_ = []

def vowel_count(word):
    count = word.count('a') + word.count('e') + word.count('i') + word.count('o') + word.count('u')
    # print(count)
    return count

for x in word_count:
    tmp_sum = vowel_count(x)
    if sum_ < tmp_sum:
        strs_.clear()
        strs_.append(x)
        sum_ = tmp_sum
    elif sum_ == tmp_sum:
        strs_.append(x)

print(sum_, strs_)
```

```python
import math

a = 1
b = 3
c = 2

def quad_sol(a, b, c):
    if b**2 - 4*a*c < 0:
        print("서로 다른 두 허근")
        return
    elif b**2 - 4*a*c == 0:
        print("중근")
    elif b**2 - 4*a*c > 0:
        print("서로 다른 두 실근")

    x1 = (-b + math.sqrt(b**2 - 4*a*c))/(2*a)
    x2 = (-b - math.sqrt(b**2 - 4*a*c))/(2*a)
    print("x1 : ", x1, "\n", "x2 : ", x2)

quad_sol(a,b,c)
quad_sol(1, -3, 2)
quad_sol(1, -2, 1)
quad_sol(1, -1, 1)
```

```python
# name = input("이름을 입력하시오")

def greet(name):
    if name == "장":
        print("그것은 저의 이름입니다. 안녕하세요")
    elif name == "Tom":
        print("안녕하세요 Tom. 반갑습니다.")
    else:
        print(name + ",안녕하세요 방가링~")
    
greet("Tom")
greet("장")
greet("Keven")
```
